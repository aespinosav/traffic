```python
def F_SO(d, a, b, am, bm):
    return 2*(d/2 + (a + am - b*d + d - 1)/(2*(b + 2*bm + 1)))*(a + b*(d/2 + (a + am - b*d + d - 1)/(2*(b + 2*bm + 1))) + d/2 + 1 - (a + am - b*d + d - 1)/(2*(b + 2*bm + 1))) - (a + am - b*d + d - 1)*(2*a + am - bm*(a + am - b*d + d - 1)/(b + 2*bm + 1) + d - (a + am - b*d + d - 1)/(b + 2*bm + 1))/(b + 2*bm + 1)
```

```python
def dc_SO(a, am , b):
    return (1 - a -am)/(1 - b)
```

```python
def F_UE(d, a, b, am, bm):
    return (-4*bm - d*(b - 1)/(b - 2*bm - 3) - 6 + 2*(a + am - 1)/b)*(2*a + am + bm*(-4*bm - d*(b - 1)/(b - 2*bm - 3) - 6 + 2*(a + am - 1)/b) - 4*bm - d*(b - 1)/(b - 2*bm - 3) + d - 6 + 2*(a + am - 1)/b) + 2*(2*bm + d*(b - 1)/(2*(b - 2*bm - 3)) + d/2 + 3 - (a + am - 1)/b)*(a + b*(2*bm + d*(b - 1)/(2*(b - 2*bm - 3)) + d/2 + 3 - (a + am - 1)/b) - 2*bm - d*(b - 1)/(2*(b - 2*bm - 3)) + d/2 - 2 + (a + am - 1)/b)
```

```python
def dc_UE(a, am, b):
    return 2*(1 - a -am)/(1 - b)
```
