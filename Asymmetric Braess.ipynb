Symmetric and Asymmetric Braess Networks=======================================

Symmetric Braess Network-------------------------We will first explore the behaviour of the symmetric Braess network with the tools we have written so far.Then we will compare this behavior with an assymetric Braess network. The way we will study it is to set some extra parameters for the costs on the links so we can change the assymetry in a quantitative manner to be able to draw some conclusions.

We first define a function to make the network, and write down the adjacency matrix and the cost function parameters.

```python
def make_network(adj, a_coefs, b_coefs):    """    Makes directed network from adjacency matrix and deletes the 'weight' attr.        It also gives the edges their a and b coef as attributes    """        g = nx.DiGraph(adj)        for i in range(len(g.edges())):        e = g.edges()[i]        del g[e[0]][e[1]]['weight']        g[e[0]][e[1]]['a'] = a_coefs[i]        g[e[0]][e[1]]['b'] = b_coefs[i]        return g
```

```python
adj_braess = np.array([0,1,1,0,                       0,0,1,1,                       0,0,0,1,                       0,0,0,0]).reshape(4,4)a = 0.5b = 0.5am = 0.1bm = 0.1a_coefs = [a, 1, am, 1, a]b_coefs = [1, b, bm, b, 1]g = make_network(adj_braess, a_coefs, b_coefs)
```

The resulting network is the familiar symmetric Braess network:

```python
plot_graph_for_flows(g)
```

We can explore the cost coefficients by analysing the parameters associated with the edges of the network. The cost functions are of the form:$$f(x) = a + bx$$ Where x is the flow on the edge.

```python
g.edges(data=True)
```

We now set a demand range for which we want to solve the static traffic assignment problem, and run the solver for the user equlibrium regime (UE):

```python
D = np.arange(0,2,0.001)sols = ta_solve_network(g, D, "UE")
```

The corresponding flows can be seen in the figure below. We can see the characteristic switching off of the middle link (in this case, link 3) and the switching on of the outer edges that are originally inactive.

```python
plot_flows(D, sols, legend=True)
```

While it is clear what is happening on the edges from the flows, once we analyse larger networks, it will not be as clear. To visualise the values of demand at which the links swith from active to inactive or *vice-versa* it is easier to plot it in a different way.In the figure below, for each edge (on the y-axis) we color in whether the link is active or inactive for the range of demand for which we solved the problem.

Asymmetric Braess network-------------------------We will now study an asymetric Braess network. For the symmetric network, like the one above, there are 2 "symmetries" that we can identify:- Links 1 and 5 are identical- Links 2 and 4 are identicalSo we can break the symmetries in 2 ways. We expect the more interesting behaviour to occur when all symmeteries are broken. But to properly understand what is happening,we will do it systematically for one of the symmetries and then for the other.For the short and easily congested roads (edges 1 and 5) we introduce the parameters $\alpha$ and $\beta$ to alter the cost functions in the following manner:$$f_1(x) = \left(a +  \dfrac{\alpha}{2} \right) + \left(1 + \dfrac{\beta}{2} \right) x$$$$f_5(x) = \left(a -  \dfrac{\alpha}{2}\right) + \left(1 - \dfrac{\beta}{2} \right) x$$By doing this we break the symmetry in a way that changes both links with the same parametrs, such that the cost of the middle route remains unchanged, and we do so in a way that doesnot give prefference to any one of the links when we allow for the parameters to take negative values as well. The values that both parameters can take will be in the range,$-1 \leq \alpha, \beta \leq 1$.Similarly we introduce parameters $\xi$ and $\eta$ to play the same role for the long and congestion resistant roads (edges 2 and 4), so that their costs are given now by:$$f_2(x) = \left(1 +  \dfrac{\xi}{2}\right) + \left(b + \dfrac{\eta}{2}\right) x \\$$$$f_4(x) = \left(1 -  \dfrac{\xi}{2}\right) + \left(b - \dfrac{\eta}{2}\right) x$$Breaking the symmetry in this way allows us to tune the parameters to adjust either the free flow costs of the links or the congestion dependent cost of them.

We will define a function that creates an assymetric braess network:

```python
def make_asym_braess(alpha, beta, xi, eta, a=0.5, b=0.5, am=0.1, bm=0.1):        adj = np.array([0,1,1,0,                    0,0,1,1,                    0,0,0,1,                    0,0,0,0]).reshape(4,4)        a = 0.5    b = 0.5    am = 0.1    bm = 0.1    a_coefs = [a + alpha/2.0, 1 + xi/2.0 , am, 1 - xi/2.0, a - alpha/2.0]    b_coefs = [1 + beta/2.0 , b + eta/2.0, bm, b - eta/2.0, 1 - beta/2.0]        return make_network(adj, a_coefs, b_coefs)
```

We will first scan through the parameters individually in a coarse interval. The demand will be in the interval $d \in [0,5]$ and the steps will be of $0.1$.The first parameters to be investigated will be $\alpha$ and $\beta$, which change the costs of the short and congestion prone links.

```python
D = np.arange(0,5.01, 0.01)alpha_range = np.arange(-1,1.2,0.2)g_alpha = [make_asym_braess(alpha, 0, 0, 0) for alpha in alpha_range]
```

```python
sols_alpha = [ta_solve_network(g_a, D, "UE") for g_a in g_alpha]
```

```python
for i in range(len(g_alpha)):    plot_flows(D, sols_alpha[i])
```

```python
for i in range(len(sols_alpha)):    plot_swithces(g_alpha[i],D, sols_alpha[i] )
```

```python
beta_range = np.arange(-1,1.2,0.2)g_beta = [make_asym_braess(0, beta, 0, 0) for beta in beta_range]sols_beta = [ta_solve_network(g_b, D, "UE") for g_b in g_beta]
```

```python
for i in range(len(g_beta)):    plot_flows(D, sols_beta[i])
```

```python
for i in range(len(sols_beta)):    plot_swithces(g_beta[i],D, sols_beta[i] )
```

Now we will scan througth the parameters $\xi$ and $\eta$ that play the equivalent role as $\alpha$ and $\beta$ in the longer *wider* links.We expect the behaviour to be different since these are the links that start off with no flow in the symmetric case. By changing the cost functions of these links we might be able to change the switching behaviour.

```python
xi_range = np.arange(-1,1.2,0.2)g_xi = [make_asym_braess(0, 0, xi, 0) for xi in xi_range]sols_xi = [ta_solve_network(g_x, D, "UE") for g_x in g_xi]
```

```python
for i in range(len(g_xi)):    plot_flows(D, sols_xi[i])
```

```python
for i in range(len(sols_xi)):    plot_swithces(g_xi[i],D, sols_xi[i] )
```

```python
eta_range = np.arange(-1,1.2,0.2)g_eta = [make_asym_braess(0, 0, eta, 0) for eta in eta_range]sols_eta = [ta_solve_network(g_e, D, "UE") for g_e in g_eta]
```

```python
for i in range(len(g_eta)):    plot_flows(D, sols_eta[i])
```

```python
for i in range(len(sols_eta)):    plot_swithces(g_eta[i],D, sols_eta[i] )
```

What we can see from changing the parameters and analysing the flows and switches is that only the $\beta$ parameter changes the gradient of the flows as functions of demand.All the other parameters just change the switching values which has the effect of making one or the other outer routes prefferable that the other. This is an expected result as the parameters reduce the the cost of the routes assymetrically.The main effect on the switching is changing the switching demand and making the switches on the outerroutes non simultaneous. The overall demand regimes of the braess network mostly remains unchanged. In some cases we do observe *off-on-off* switches that are originally not present in the symmetric braess network.
